```python
import cv2
import matplotlib.pyplot as plt
import pandas as pd
```

# EDA
Read preprocessed csv file.

```python
train = pd.read_csv("../input/train.csv")
```

```python
train.head()
```

```python
print(len(train))
```

```python
print(train["target"].value_counts() / len(train) * 100)
```

## Class Distribution

The Label Encoding is as follows:

| Target | Value |
|---------|-------|
|0| Red Blood Cell |
|1| Ring |
|2| Schizont |
|3| Trophozite |


```python
train["target"].value_counts().plot(kind="bar", rot=0, title="Train Distribution of Target",
                                    figsize=(10, 7), xlabel = "Blood Cell Category", ylabel="Count")
plt.xticks(ticks=[0, 1, 2, 3], labels=["Red Blood Cell", "Trophozite", "Ring", "Schizont"])
plt.show()
```

As we can see, there is **severe class imbalance** among the images.

## View images

Let us plot a few images for each target value

### 1. Red Blood Cells

```python
rbc_ids = list(train.loc[train["target"] == 0, "image_id"][:5])

fig = plt.figure(figsize=(20, 5))
for i, rbc_id in enumerate(rbc_ids):
    a = fig.add_subplot(1, 5, i + 1)
    img = cv2.imread(f"../input/train/{rbc_id}.png")
    plt.imshow(img)
    plt.title(rbc_id)

plt.suptitle("Red Blood Cells", size=20)
plt.tight_layout()
plt.show()
```

### 2. Ring

```python
ring_ids = list(train.loc[train["target"] == 1, "image_id"][:5])

fig = plt.figure(figsize=(20, 5))
for i, ring_id in enumerate(ring_ids):
    a = fig.add_subplot(1, 5, i + 1)
    img = cv2.imread(f"../input/train/{ring_id}.png")
    plt.imshow(img)
    plt.title(rbc_id)

plt.suptitle("Ring Stage Infected Cells", size=20)
plt.tight_layout()
plt.show()
```

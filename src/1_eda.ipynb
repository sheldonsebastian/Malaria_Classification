```python
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

# EDA
Read preprocessed csv file.

```python
train = pd.read_csv("../input/train.csv")
holdout = pd.read_csv("../input/holdout.csv")
```

Target column represents

|Target| Value |
|--------------|-------|
|0| Red Blood Cell |
|1| Ring |
|2| Schizont |
|3| Trophozoite |

The distribution of target in train and holdout set are similar.

## Class Distribution

```python
train["target"].value_counts().plot(kind="bar", rot=0, title="Train Distribution of Target", figsize=(10, 7))
plt.show()
```

As we can see, there is **severe class imbalance** among the images.

## View images

Let us plot a few images for each target value

### 1. Red Blood Cells

```python
rbc_ids = list(train.loc[train["target"]==0, "image_id"][:5])

fig = plt.figure(figsize=(20, 5))
for i, rbc_id in enumerate(rbc_ids):
    a = fig.add_subplot(1, 5, i+1)
    img = cv2.imread(f"../input/train/{rbc_id}.png")
    plt.imshow(img)
    plt.title(rbc_id)
    
plt.suptitle("Red Blood Cells", size=20)
plt.tight_layout()
plt.show()
```

### 2. Ring

```python
ring_ids = list(train.loc[train["target"]==1, "image_id"][:5])

fig = plt.figure(figsize=(20, 5))
for i, ring_id in enumerate(ring_ids):
    a = fig.add_subplot(1, 5, i+1)
    img = cv2.imread(f"../input/train/{ring_id}.png")
    plt.imshow(img)
    plt.title(rbc_id)
    
plt.suptitle("Ring Infected Cells", size=20)
plt.tight_layout()
plt.show()
```
